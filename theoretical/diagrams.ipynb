```python
import matplotlib.pyplot as plt
```

```python
M = 10
def radiusratio(i,M):
    return (M-i)/(M-i+1)

ids = list(range(1,M))
rr = list(map(lambda i : radiusratio(i,M), ids))
```

```python
plt.plot(ids,rr,'o-')
plt.title('Radius Ratio')
plt.ylim(0.4,1)
plt.xlabel('Rank')
plt.ylabel('Radius ratio with next rank')
plt.show()
```

```python
M = 10
def radiusratio(i,M):
    return (i-1)/(i)

ids = list(range(2,M+1))
rr = list(map(lambda i : radiusratio(i,M), ids))
```

```python
plt.plot(ids,rr)
plt.show()
```

```python
M = 10
I0 = 1
p0 = 0.8
def Irecu(i):
    return I0*pow(p0,i-1)*pow((M-i+1)/M,2)

def Iabs(i):
    #if i == 1: return I0*(1-p0)
    return I0*pow(p0,i-1)*(1-p0)*pow((M-i+1)/M,2)


def Itr(i):
    return I0*pow(p0,i)*(2*M+1-2*i)/(M**2) # *(1-pow((M-i)/M,2))

ids = list(range(1,M+1))
irec = list(map(Irecu, ids))
iabs = list(map(Iabs, ids))
itr = list(map(Itr, ids))
```

```python
Iabs(1)
```

```python
#plt.plot(ids,irecall)
plt.plot(ids,irec,'o-',label='Ei received by the upper part')
plt.plot(ids,itr,'o-',label='Ei transmitted to lower part')
plt.plot(ids,iabs,'o-',label='Ei absorbed by upper part')
plt.legend()
plt.xlim(0,M+1)
plt.xticks(ids)
plt.xlabel('Rank')
plt.ylabel('Energy')
plt.show()
```

```python
for i in range(M-1):
    input = irec[i]
    output = irec[i+1]+iabs[i]+itr[i]
    print(ids[i], input, output, abs(input-output)) #, irec[i+1],iabs[i],itr[i])
```
